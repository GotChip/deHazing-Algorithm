```python
import PIL.Image as Image
import skimage.io as io
import numpy as np
import matplotlib.pyplot as plt
import os
import shutil
from tqdm.notebook import tqdm
import torchvision
from torchvision.io import read_image
import torch
from torch import tensor
import torchmetrics.regression
from torchmetrics.regression import MeanSquaredError
```

```python
def eval(output, groundTruth):
  groundTruth = read_image(groundTruth) # read the images
  output = read_image(output)

  # Initialize a list to store MSE for each slice
  mse_list = []

  # Loop over slices of the 3D tensors
  for i in range(output.shape[0]):
    for j in range(output.shape[1]):
      for k in range(output.shape[2]):
        slice1 = output[i,j,k]
        slice2 = groundTruth[i,j,k]

      # Calculate MSE for the slices
      mean_squared_error = MeanSquaredError(num_outputs=1)
      mse = mean_squared_error(slice1, slice2)

      # Append the MSE to the list
      mse_list.append(mse.item())

  # Print the MSE for each slice
  for i, mse in enumerate(mse_list):
    amse = mse

  amse = amse / (output.shape[0] *output.shape[1] *output.shape[2])

  return amse
```

```python
from pathlib import Path
directory = r"/content/drive/MyDrive/hazeOutput"
mselist = []

for name in tqdm(os.listdir(directory)):
  array1 = name.split('_')
  if not array1[-1] == '.ipynb_checkpoints':
    with open(os.path.join(directory, name)) as f:

      path = directory + "/" + name

      array1 = path.split('_')
      array2 = array1[0].split('/')
      array3 = array2[-1].split('o')
      img_name = "/content/drive/MyDrive/clear/" + array3[0] +".jpg"

      mse = eval(path, img_name)

      mselist.append(mse)
```
